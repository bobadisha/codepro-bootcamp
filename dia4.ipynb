```python
class Personajes:
    tipo = "ficcion"

    def __init__(self, nombre, poder, frase):
        self.nombre = nombre
        self.poder = poder
        self.frase = frase

    def info_personaje(self):
        print(f"Mi nombre es {self.nombre}, mi poder es {self.poder} y digo: {self.frase}")


class Heroe(Personajes):
    pass


class Villano(Personajes):
    pass


personaje1 = Heroe("Superman", "Volar, Super Fuerza, etc", "Soy Superman")
personaje2 = Villano("Lex Luthor", "Intelecto", "No necesito poderes… solo necesito ser más listo")
personaje3 = Heroe("Groot", "Fuerza, Regeneración", "Yo soy Groot")

personaje1.info_personaje()
personaje2.info_personaje()
personaje3.info_personaje()
```
