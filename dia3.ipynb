```python
'''Challenge 4 ず - 20 min
Crear funci贸n que revise si una palabra contiene una letra. 
En caso de que la contenga, retornar la letra entre corchetes []. 
De lo contrario, retornar solo la letra.'''

#crear variables para ingresar datos 
#variable = input etc
#letra_buscada = 'a'


palabra_random = input('Ingrese una palabra')
letra_buscada = 'a'

#definir una funcion para analizar la palabra 
#comparar con If , si (if) la letra que se busca (letra_buscada) esta en (in) la frase escrita (palabra_frandom)
#que retorne (return) entre corchete [] la letra_buscada 
#si no contiene (else) que retorne sin corchete []
#ultimo llamar a la funcion con la palabra_random
def analizar_palabra(palabra_random):
    if letra_buscada in palabra_random:
        return [letra_buscada ]
    else:
        return letra_buscada 
    
analizar_palabra(palabra_random)
```
